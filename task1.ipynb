```python
from transformers import pipeline

def summarize_text(text, max_length=150, min_length=50):
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)
    return summary[0]['summary_text']

if __name__ == "__main__":
    print("Enter the text you want to summarize (Press Enter twice to finish):")
    user_input = []

    while True:
        try:
            line = input()
            if line.strip() == "":
                break
            user_input.append(line)
        except EOFError:
            break

    input_text = " ".join(user_input)

    if len(input_text) < 50:
        print("Text is too short for summarization. Please enter a longer text.")
    else:
        summarized_text = summarize_text(input_text)
        print("\n--- Summarized Text ---\n")
        print(summarized_text)
```

```python
import gradio as gr
from transformers import pipeline

def summarize_text(text, max_length=150, min_length=50):
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)
    return summary[0]['summary_text']

iface = gr.Interface(
    fn=summarize_text,
    inputs=[
        gr.Textbox(lines=5, placeholder="Enter text to summarize..."),
        gr.Slider(50, 500, value=150, label="Max Summary Length"),
        gr.Slider(20, 100, value=50, label="Min Summary Length")
    ],
    outputs="text",
    title="Text Summarization Tool",
    description="Enter a paragraph and get a summarized version using a pre-trained BART model."
)

iface.launch()
```
